# **WINE QUALITY PREDICTION** USING **LINEAR REGRESSION**

## **Import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn import metrics
from sklearn.linear_model import LinearRegression

import warnings
warnings.filterwarnings('ignore')
```

## **Reading the data**

```python
df = pd.read_csv('winequality.csv')
print(df.head(2))
```

```python
df.describe()
```

## **Checking the Null Values**

```python
df.isnull().sum()
```

```python
for col in df.columns:
  if df[col].isnull().sum() > 0:
    df[col] = df[col].fillna(df[col].mean())

df.isnull().sum().sum()
```

## **Visualization**

```python
df.hist(bins=10, figsize=(10, 10))
plt.show()
```

```python
plt.bar(df['quality'], df['alcohol'])
plt.xlabel('quality of alcohol')
plt.ylabel('alcohol')
plt.show()
```

```python
plt.figure(figsize=(5, 4))
sb.heatmap(df.corr() > 0.7)
plt.show()
```

```python
df['best quality'] = [1 if x > 5 else 0 for x in df.quality]
```

## **Data Splitting**

```python
features = df.drop(['quality', 'best quality'], axis=1)
target = df['best quality']
xtrain, xtest, ytrain, ytest = train_test_split(features, target, test_size=0.2, random_state=40)
xtrain.shape, xtest.shape
```

## **Scaling the Data**

```python
norm = MinMaxScaler()
xtrain = norm.fit_transform(xtrain)
xtest = norm.transform(xtest)
```

## **Training , Testing and Determining the model performance**

```python
lr= LinearRegression()
lr.fit(xtrain,ytrain)
print('Training Accuracy : ', metrics.roc_auc_score(ytrain, lr.predict(xtrain)))
print('Validation Accuracy : ', metrics.roc_auc_score(ytest, lr.predict(xtest)))
```
